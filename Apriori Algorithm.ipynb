## Preprocessing

```python
import pandas as pd
data = pd.read_csv("Patient_Characteristics_Survey__PCS___2019.csv")
pd.set_option('display.max_columns', None)
data.head()
```

```python
def f(row):
    if row['Hyperlipidemia'] == "YES" or row['High Blood Pressure'] == "YES" or row['Diabetes'] == "YES" or row['Obesity'] == "YES" or row['Heart Attack']	== "YES" or row['Stroke']	== "YES" or row['Other Cardiac'] == "YES" or row['Pulmonary Asthma']	== "YES" or row['Alzheimer or Dementia'] == "YES" or row['Kidney Disease'] == "YES" or row['Liver Disease'] == "YES" or row['Endocrine Condition'] == "YES" or row['Neurological Condition'] == "YES" or row['Traumatic Brain Injury'] == "YES" or row['Joint Disease'] == "YES" or row['Cancer']	== "YES" or row['Other Chronic Med Condition'] == "YES": 
        val = "YES"
    elif row['Hyperlipidemia'] == "NO":
        val = 'NO'
    else:
        val = 'Unknown'
    return val

#create new column 'Chronic Illness' using the function above
data['Chronic Illness'] = data.apply(f, axis=1)
```

```python
def f(row):
    if row['Alcohol Related Disorder'] == "YES" or row['Drug Substance Disorder'] == "YES" or row['Opioid Related Disorder'] == "YES":
        val = 'YES'
    elif row['Alcohol Related Disorder'] == "NO":
        val = 'NO'
    else:
        val = 'Unknown'
    return val

#create new column 'Substance Abuse' using the function above
data['Substance Abuse'] = data.apply(f, axis=1)
```

```python
df = pd.DataFrame(data, columns=['Age Group', 'Sex', 'Chronic Illness', 'Substance Abuse','Mental Illness','Criminal Justice Status'])
df
```

```python
df.isnull().sum()
```

## Apriori Algorithm

```python
df['Chronic Illness'] = df['Chronic Illness'].replace(['YES'],'CI-YES')
df['Chronic Illness'] = df['Chronic Illness'].replace(['NO'],'CI-NO')
df['Chronic Illness'] = df['Chronic Illness'].replace(['Unknown'],'CI-UN')
df['Substance Abuse'] = df['Substance Abuse'].replace(['YES'],'SA-YES')
df['Substance Abuse'] = df['Substance Abuse'].replace(['NO'],'SA-NO')
df['Substance Abuse'] = df['Substance Abuse'].replace(['Unknown'],'SA-UN')
df['Mental Illness'] = df['Mental Illness'].replace(['YES'],'MI-YES')
df['Mental Illness'] = df['Mental Illness'].replace(['NO'],'MI-NO')
df['Mental Illness'] = df['Mental Illness'].replace(['Unknown'],'MI-UN')
df['Criminal Justice Status'] = df['Criminal Justice Status'].replace(['YES'],'CJ-YES')
df['Criminal Justice Status'] = df['Criminal Justice Status'].replace(['NO'],'CJ-NO')
df['Criminal Justice Status'] = df['Criminal Justice Status'].replace(['Unknown'],'CJ-UN')
df
```

```python
dataList = df.values.tolist()
```

```python
from apyori import apriori
```

```python
association_rules = apriori(dataList, min_support = 0.2, min_confidence = 0.2, min_lift = 0.2)
association_results = list(association_rules)
association_results
```

```python
print(len(association_results))
```

```python
df = pd.DataFrame(columns=('Items','Antecedent','Consequent','Support','Confidence','Lift'))

Support =[]
Confidence = []
Lift = []
Items = []
Antecedent = []
Consequent=[]

for RelationRecord in association_results:
    for ordered_stat in RelationRecord.ordered_statistics:
        Support.append(RelationRecord.support)
        Items.append(RelationRecord.items)
        Antecedent.append(ordered_stat.items_base)
        Consequent.append(ordered_stat.items_add)
        Confidence.append(ordered_stat.confidence)
        Lift.append(ordered_stat.lift)

df['Items'] = list(map(set, Items))                                   
df['Antecedent'] = list(map(set, Antecedent))
df['Consequent'] = list(map(set, Consequent))
df['Support'] = Support
df['Confidence'] = Confidence
df['Lift']= Lift
```

```python
df.sort_values(by ='Confidence', ascending = False, inplace = True)
df
```
